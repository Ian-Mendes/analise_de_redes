###**Questão 1**
Defina:

(a) Subgraph:

  Subgrafos são "partes" de um grafo maior. Dado um Grafo A, um subgrafo B de A é um grafo cujos vértices também são vértices de A, assim como as arestas de B também são arestas de A.

(b) Bipartite graph:

  Grafos com os nós divididos em dois conjuntos separados. Os nós de um mesmo conjunto não possuem arestas ligando-os, de forma que cada conjunto separado de nós apenas se conecta ao outro conjunto. Por isso utiliza-se o termo "bipartido".
  
  ![Grafos Bipartidos](https://mathworld.wolfram.com/images/eps-svg/BipartiteGraph_1000.svg)


(c) Hamiltonian graph:

  Um grafo Hamiltoniano é um grafo que possua um circuico Hamiltoniano. Um circuito Hamiltoniano é um circuito no qual é possível percorrer caminhos por todos os nós sem repetir nó algum, além do inicial e final.

(d) Eulerian graph:

  Grafo cujos nós possuem grau par. Precisa incluir trilha fechada que que inclua todas as arestas do grafo.

###**Questão 2**
Describe how a breadth-first search algorithm works.

Breadth-First Search (BFS) é um algoritmo de busca em grafos, no qual os nós já visitados são assinalados para fins de acompanhamento. Após a escolha do nó inicial, analisam-se caminhos a partir dos nós vizinhos a esse, assinalando-os e repetindo o processo de análise de vizinhos dos novos nós.
Breadth vem do inglês e significa _"amplitude"_ ou _"largura"_. Como o nome sugere, a preferência é dada à análise de nós em uma mesma camada, prosseguindo para as próximas em conexão com os nós em questão.

### **Questão 3**
How many edges does a complete graph with n vertices have(I)? What about a complete directed graph with n vertices? (II)

(I) Utilizando combinação, podemos encontrar a resposta. Ora, se o grafo é completo, então há arestas em todas as combinações possíveis de vértices.
$$C_{n,2} = \frac{n!}{2!(n-2)!} = \frac{n(n-1)}{2}$$

(II) Em grafos direcionados, ao contrário do visto no item (I), as arestas são bidimensionais. Revisando a fórmula, temos que:
$$2*\frac{n(n-1)}{2} = n*(n-1)$$

* Mesmo número de vértices;
* Mesmo número de arestas;

### Questão 6
Given the digraph G = (V, E) where V = {M, N, O, P, Q, R, S} and

E ={(M, S), (N, O), (P, R), (N, S), (O, M), (N, Q), (O, M), (P, P), (S, M), (O, N),  (S, M), (N, R), (P, M), (M, S)}

(a) Specify, if any, a simple path from vertex M to vertex S.

(b) Specify, if any, a simple cycle, involving at least 4 nodes.

(c) Is the digraph connected or not connected?

(d) What is the degree of vertices N and R.

(e) Represent the digraph using adjacency list representation.

(f) Represent the digraph using adjacency matrix representation.


````python
import networkx as nx
import matplotlib as plt
from networkx import DiGraph

#criando o Dígrafo
G = nx.DiGraph()
#adicionando nós
G.add_nodes_from(['M','N','O','P','Q','R','S'])
#adicionando arestas
G.add_edges_from([('M','S'), ('N', 'O'), ('P', 'R'), ('N', 'S'), ('O', 'M'), ('N', 'Q'), ('O', 'M'), ('P', 'P'), ('S', 'M'), ('O', 'N'), ('S', 'M'), ('N', 'R'), ('P', 'M'), ('M', 'S')])
````

Item (a): Especifique, caso exista, um caminho de 'M' a 'S'

````python
#Há caminho entre os nós M e S?
#Saída: resposta booleana para a questão acima
print(nx.is_path(G,['M','S']))
#Há caminho simples entre os nós M e S? (Sem auto loops)
#Saída: resposta booleana para a questão acima.
print(nx.is_simple_path(G,['M','S']))
````

Item (b): Specify, if any, a simple cycle, involving at least 4 nodes.

````python
#Lista de ciclos simples!
print(list(nx.simple_cycles(G)))
#Ao printarmos a lista, percebemos que não há ciclos simples com 4 nós
````

````python
#Vamos visualizar o Grafo?
nx.draw(G,with_labels=True)
````

Item (c): Is the digraph connected or not connected?

````python
#"A graph is semiconnected if and only if for any pair of nodes, either one is reachable from the other, or they are mutually reachable."
#Como podemos ver acima, Q não é capaz de alcançar nenhum outro nó. Por contraexemplo, já é possível constatar que o retorno será falso.
nx.is_semiconnected(G)
````

Item (d):What is the degree of vertices N and R.

````python
#Quais são os graus dos nós 'N' e 'R'
nx.degree(G,{"N","R"})
````

Item (e): Represent the digraph using adjacency list representation.

````python
# Listas de adjacências representam em formato de String quais nós são adjacentes ao nó especificado. O primeiro nó, em cada lista, é o nó de origem. Os nós que se seguem são aqueles adjacentes a ele.
# Em nosso exemplo, temos que a adjacência de M é apenas o nó S (os dois são direcionados um ao outro).
# O nó P possui adjacência com R, P (ele mesmo) e M.
list(nx.generate_adjlist(G))
````

item (f): Represent the digraph using adjacency matrix representation.

````python
#DÚVIDA
A = nx.adjacency_matrix(G,nodelist=sorted(G.nodes()))
print(A.todense())
````
